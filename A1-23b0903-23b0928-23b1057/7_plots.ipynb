```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#violin plot
np.random.seed(10)
data = np.random.normal(80,20,200)

plt.violinplot(data)

plt.title("Violin Plot for Single Collection")
plt.xlabel("Category")
plt.ylabel("Values")

plt.show()
```

```python
#pareto chart
flavors = ['Chocolate', 'Vanilla', 'Butterscotch', 'BlackCurrent', 'Strawberry']
orders = [1500,  670,  950,  450,   75]

sorted = np.argsort(orders)[::-1]
flavors = np.array(flavors)[sorted]
orders = np.array(orders)[sorted]

cumulative_orders = np.cumsum(orders)
total = np.sum(orders)
percentage = (cumulative_orders/total)*100

fig, ax1 = plt.subplots()

ax1.bar(flavors, orders, color='b')
ax1.set_xlabel('Flavors')
ax1.set_ylabel('Orders', color='b')

ax2 = ax1.twinx()
ax2.plot(flavors, percentage, 'o-', color='r')
ax2.set_ylabel('Percentage', color='r')

plt.title("Pareto Chart for Ice Cream Flavors")

plt.show()
```

```python
#coxcomb chart
flavors = ['Chocolate', 'Vanilla', 'Butterscotch', 'BlackCurrent', 'Strawberry']
orders = [1000,  670,  850,  450,   375]

angles = np.linspace(0, 2 * np.pi, len(flavors), endpoint=False).tolist()

fig, ax = plt.subplots(figsize = (6,6) , subplot_kw={'projection': 'polar'})

colors = ['red', 'blue', 'green', 'yellow', 'cyan']

bars = ax.bar(angles, orders, width=1.2, bottom=0, color = colors, edgecolor='black')

ax.set_xticks(angles)
ax.set_xticklabels(flavors)

ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
```

```python
x = np.linspace(0, 10, 100)  
y = np.linspace(0, 5, 50)    
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Waterfall Plot')

plt.show()
```
