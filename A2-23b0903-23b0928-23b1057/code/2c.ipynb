```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#function that samples from the gaussian distribution with mean = loc and std deviation = scale
def sample(loc, scale):
    uniform_random_numbers = np.random.uniform(0, 1, 100000)
    X = norm.ppf(uniform_random_numbers, loc=loc, scale=scale)
    return X
```

```python
#generating 10^5 samples for four parameter choices
X1 = sample(0, np.sqrt(0.2))
X2 = sample(0, np.sqrt(1))
X3 = sample(0, np.sqrt(5))
X4 = sample(-2, np.sqrt(0.5))
```

```python
#ploting the samples for each parameter choice
X1_hist = plt.hist(X1, bins=100, density=True, alpha=0.5, label=r"$\mu=0, \sigma^2=0.2$", color='blue')
X2_hist = plt.hist(X2, bins=100, density=True, alpha=0.5, label=r"$\mu=0, \sigma^2=1$", color='red')
X3_hist = plt.hist(X3, bins=100, density=True, alpha=0.5, label=r"$\mu=0, \sigma^2=5$", color='yellow')
X4_hist = plt.hist(X4, bins=100, density=True, alpha=0.5, label=r"$\mu=-2, \sigma^2=0.5$", color='green')
plt.xlim(-6, 6)
plt.xlabel("X")
plt.ylabel("p(X)")
plt.legend()
plt.show()
```
