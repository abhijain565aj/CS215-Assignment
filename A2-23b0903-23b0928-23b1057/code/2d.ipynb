```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# # sampling normal distributions using a Galton board

# def Galton_board(N,h):
    # pockets = np.zeros(2*h+1)
    # for i in range(N):
    #     x = 0
    #     for j in range(h):
    #         move = np.random.choice([-1,1], p=[0.5,0.5]) #-1 for left, 1 for right
    #         x += move       
    #     pockets[x+h] += 1    
    # return pockets

def Galton_board(N, h):
    moves = np.random.choice([-1, 1], size=(N, h))
    final_positions = np.sum(moves, axis=1)
    shifted_positions = final_positions + h
    pockets = np.bincount(shifted_positions, minlength=2*h+1)
    
    return pockets
```

```python
#plotting the count of balls in each pocket
p1 = Galton_board(100000,10)
plt.hist(range(-10,11), weights=p1, bins=21)
plt.xlabel('Pocket')
plt.ylabel('Count of balls')
plt.title('Galton board with depth 10')
plt.show()
```

```python
p2 = Galton_board(100000,50)
plt.hist(range(-50,51), weights=p2, bins=101)
plt.xlim(-25,25)
plt.xlabel('Pocket')
plt.ylabel('Count of balls')
plt.title('Galton board with depth 50')
plt.show()
```
