```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# # sampling normal distributions using a Galton board
def Galton_board(N,h):
    pockets = np.zeros(h+1)
    for i in range(N):
        x = 0
        for j in range(h):
            move = np.random.choice([0,1], p=[0.5,0.5]) #0 for left, 1 for right
            if move == 0:
                x -= 1
            if move == 1:
                x += 1       
        x = min(x+h,h)
        pockets[x] += 1    
    return pockets
# def Galton_board(N, h):
#     # Generate movements for all balls: shape (N, h)
#     moves = np.random.choice([0, 1], size=(N, h), p=[0.5, 0.5])
    
#     # Calculate the final position for each ball
#     positions = np.sum(moves, axis=1) - (h // 2)
    
#     # Convert to pocket indices
#     pockets_indices = np.clip(positions + (h // 2), 0, h)
    
#     # Count the number of balls in each pocket
#     pockets_count = np.bincount(pockets_indices, minlength=h + 1)
    
#     # Normalize the counts to get fractions
#     # pockets_fraction = pockets_count / N
    
    # return pockets_count
```

```python
#plotting the count of balls in each pocket
p1 = Galton_board(1000,100)
plt.hist(range(-50,51), weights=p1, bins=101, width=0.5)
# plt.xlim(-40, 40)
plt.xlabel('Pocket')
plt.ylabel('Count of balls')
plt.title('Galton board with depth 10')
plt.show()
```
